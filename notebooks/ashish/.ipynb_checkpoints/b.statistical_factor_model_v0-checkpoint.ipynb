```python
import pandas as pd
import numpy as np
import os
import datetime
import matplotlib.pyplot as plt
```

### Get data


```python
df_agg = pd.read_parquet("../../data/data_agg_v0.parquet")

price_cols = [x for x in df_agg.columns if "LAST" in x]
df_subset = df_agg[["Date"] + price_cols[:12]]
df_subset.index = df_subset["Date"]
df_subset = df_subset.drop(columns=["Date"])
print(f"Shape before dropping nulls, {df_subset.shape}")
df_subset = df_subset.dropna()
print(f"Shape after dropping nulls, {df_subset.shape}")
```

```python
df_subset.plot()
```

```python
df_subset1 = df_subset.sort_index()
df_subset1 = np.log(df_subset1/df_subset1.shift(1)).dropna()
df_subset1.plot()
```

### PCA

```python
X = df_subset1.values
X_mean = X.mean(axis=0)
print (X.shape,X_mean.shape)

N,M = X.shape

X_normalised = X-X_mean

cov = (X_normalised.T @ X_normalised)/(N)
plt.matshow(cov)
plt.colorbar()
plt.title("Covariance of $X$")

U,S,Vt = np.linalg.svd(X_normalised)
print ("Singular values")
print(U.shape,S.shape,Vt.shape)

e,e_vec=np.linalg.eig(cov)
print ("Eigen values")
print(e.shape,e_vec.shape)
```

```python
plt.matshow(df_subset1.corr())
plt.colorbar()
```

```python
plt.semilogy(S**2/N,"-.", label = "Singular values")
plt.semilogy(e,"-.", label = "Eigen values")
plt.legend()
plt.title("Eigen Values")
```

```python
fig,ax=plt.subplots(1,2,figsize=(16,9))
for i in range(6):
    ax[0].plot(e_vec[:,i], label=f"Eigen Vector {i}")
    ax[1].plot(Vt.T[:,i], label=f"Eigen Vector (SVD) {i}")
ax[0].legend()
ax[1].legend()
```
